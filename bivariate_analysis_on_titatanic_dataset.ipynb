```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic.head()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'sex')
```

```python
sns.barplot(data = titanic, x = 'pclass', y = 'age', hue = 'sex')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)
sns.scatterplot(data=tips, x='total_bill', y='tip', hue = 'sex')
plt.subplot(1,2,2)
sns.barplot(data = titanic, x = 'pclass', y = 'age', hue = 'sex')
```

```python
sns.distplot(titanic[titanic.survived == 0].age, hist = False, label = 'died')
sns.distplot(titanic[titanic.survived == 1].age, hist = False, label = 'survived')
plt.legend()
plt.show()
```

```python
titanic.groupby(['pclass','sex'])['survived'].mean()
```

```python
titanic.groupby(['pclass','sex'])['survived'].mean().plot(kind = 'bar')
```

```python
titanic.groupby(['embarked'])['survived'].mean().plot(kind = 'bar')
```

```python
sns.pairplot(titanic, hue = 'survived')
```

```python
titanic.groupby(['pclass'])['survived'] .mean().plot(kind = 'pie')
```
